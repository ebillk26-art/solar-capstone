```python
# Check what's already installed
import sys

packages_to_check = {
    'ultralytics': 'Ultralytics',
    'torch': 'PyTorch',
    'torchvision': 'TorchVision',
    'PIL': 'Pillow',
    'cv2': 'OpenCV',
    'numpy': 'NumPy',
    'pandas': 'Pandas',
    'sklearn': 'Scikit-learn',
    'matplotlib': 'Matplotlib',
    'seaborn': 'Seaborn',
    'yaml': 'PyYAML',
    'tqdm': 'tqdm',
}

print("Checking installed packages...\n")
installed = []
missing = []

for module, name in packages_to_check.items():
    try:
        __import__(module)
        print(f"✅ {name} - INSTALLED")
        installed.append(name)
    except ImportError:
        print(f"❌ {name} - NOT INSTALLED")
        missing.append(name)

print("\n" + "="*50)
print(f"Installed: {len(installed)}/{len(packages_to_check)}")
print("="*50)

if not missing:
    print("\n🎉 ALL PACKAGES ALREADY INSTALLED!")
    print("You can skip installation and start training!")
else:
    print(f"\n⚠️  Missing: {', '.join(missing)}")
    print("Contact your IT to install these.")
```
